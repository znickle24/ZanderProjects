```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request
import scipy as sc
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import norm, bernoulli, binom, poisson
```

```python
newFile = open('hackerrank.html', 'w')

for p in range(1, 6): 
    #allow the url to be formatted for each page through the first 5 pages
    url = "https://news.ycombinator.com/news?p={}".format(p)
    with urllib.request.urlopen(url) as response:
        html = response.read()
        html = html.decode('utf-8')
        # save the file
        newFile.write(html)
newFile.close()
```

```python
soup = BeautifulSoup(open("hackerrank.html").read(), 'html.parser')
# print(soup.prettify())
```

```python
tables = soup.find_all("table", class_= 'itemlist')
```

```python
def stringToInt(age): 
    try: 
        if("minutes" in age):
            num = int(age.replace("minutes ago", ""))
            ret = num/60
            return ret
        if("minute" in age): 
            num = age.replace("minute ago", "")
            ret = num/60
            return ret
        if("hours" in age): 
            ret = int(age.replace("hours ago", ""))
            return ret
        if("hour" in age): 
            ret = int(age.replace("hour ago", ""))
            return ret
        if("days" in age): 
            num = int(age.replace("days ago", ""))
            ret = num*24
            return ret
        if("day" in age):
            num = int(age.replace("day ago", ""))
            ret = num*24
            return ret
        return 1
    except ValueError:
        return 1
def scoring(scoreInfo):
    if (scoreInfo is None):
        return 0
    else:
        score = scoreInfo.string
        scoreVal = int(score.replace("points", ""))
        return scoreVal
def commenting(commentText):
    if (commentText.string == 'hide' or commentText.string == 'discuss'):
        return 0
    elif ("comments" in commentText.string):
        comment = commentText.string
        commentNum = int(comment.replace("comments", ""))
        return commentNum
    else: 
        return 1
def setFirstPage(rank):
    if (rank < 31):
        return 1
    else:
        return 0
df = pd.DataFrame(columns=["Rank", "Title_Length", "Age", "Score", "Number_of_Comments", "On_First_Page"])
for table in tables:
    athings = table.find_all('tr', class_='athing')
    subtexts = table.find_all('td', class_='subtext')
    for index in range(30):
        athing = athings[index]
        subtext = subtexts[index]
        rankInfo = athing.find('span', class_='rank').string
        rank = int(rankInfo.replace(".", ""))
        
        onFirstPage = setFirstPage(rank)
        title = athing.find('a', class_='storylink').string
        titleLength = int(len(title))
        ageInfo = subtext.find('span', class_='age')
        ageString = ageInfo.find('a').string
        age = float(stringToInt(ageString))
        scoreInfo = subtext.find('span', class_='score')
        score = int(scoring(scoreInfo))
        c = subtext.find_all('a')
        commentText = c[-1]
        comments = commenting(commentText)
        df = df.append({"Rank" : rank, "Title_Length" : titleLength, 
                        "Age" : age, "Score" : score, "Number_of_Comments" : comments, "On_First_Page": onFirstPage}, ignore_index=True)
#         print("Comments: ", comments)    
#         print("Score: ", score)
#         print("Age: ", age)
#         print("Rank: ", rank)
#         print("Title: ", titleLength)
display(df)
```

```python
compareAll = sm.ols(formula="Rank ~ Age + Title_Length + Score + Number_of_Comments", data=df).fit()
compareAll.summary()
```

```python
compareAge = sm.ols(formula="Rank ~ Age", data=df).fit()
compareAge.summary()
```

```python
compareScore = sm.ols(formula="Rank ~ Score", data=df).fit()
compareScore.summary()
```

```python
rankModel = sm.logit(formula="On_First_Page ~ Age", data=df).fit()
rankModel.summary2()
```

```python
df['Age_Rank_Prediction'] = rankModel.predict()>.5
```

```python
pd.crosstab(index=df["On_First_Page"], columns=df["Age_Rank_Prediction"]) # confusion matrix
```

```python
rankModel2 = sm.logit(formula="On_First_Page ~ Score + Age", data=df).fit()
rankModel2.summary2()
```

```python
df['Score_Age_Squared_Prediction'] = rankModel2.predict()>.5
```

```python
pd.crosstab(index=df["On_First_Page"], columns=df["Score_Age_Squared_Prediction"]) # confusion matrix
```

```python
ax = df[df['On_First_Page']==0].plot.scatter(x='Age',y='Score',color='red',label='not on first page')
df[df['On_First_Page']==1].plot.scatter(x='Age',y='Score',color='blue',label='on first page', ax=ax)

x = sc.linspace(df.Age.min(),df.Age.max(),1000)
p = sc.linspace(1e-5,1-1e-5,1000)
odds = lambda p: p/(1-p)
logit = lambda p: sc.log(odds(p))
par = dict(rankModel2.params)
# plt.plot(x,(1/par['On_First_Page'])*(logit(.5) - par['Score'] - par['Age']*x),color='Black')
plt.plot(x, (1/par['Score'])*(logit(.5) - par['Intercept'] - par['Age']*x), color='Black')
```

*
From the information in the confusion matrix, it would appear that score and age are decent predictors of making the front page or not. In the model above, you can see that it did fail a number of times in predicting, but for those on the front page, it predicted that a decent number of times.
